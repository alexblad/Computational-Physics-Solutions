```python
from pylab import *
import numpy as np

#Loads in data from text file
data=np.loadtxt('sunspots.txt')

#Assigns data to a variable
month, sunid=data[:,0], data[:,1]

#Creates plot with month as the x-axis and sunid as the y-axis
plot(month, sunid)
show()

```

```python
#Create plot using only the first 1000 data points
#Assigns data to a variable
month, sunid=data[:1000,0], data[:1000,1]

#Create plot
plot(month, sunid, 'g')
show()

```

```python
#Find the running average

r=5

#Create an array of the data points for the running average
s = [ sum(sunid[k-r:k+r]) for k in range(1000)]

#Running average
y=array(s)/2/r

plot(month,sunid,'g.',label='noisy')
plot(month,y,label='averaged')
legend()
show()
```
